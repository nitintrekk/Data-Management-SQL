```python


# import random package
import random 
# Ask the user name
user_name = input("Hi, and welcome to Python! Whats your name? ").capitalize()
# Ask if the user wants to play game theory
user_play = input("Do you want to play a Game Theory Situation? ").lower()


# If user want to play
if "yes" in user_play:
    
    print(f"\n{user_name} We were caught! Do not confess! We didn't do anything!")
    
    print("""
    You are in small room at the police station. You can't communicate with anyone.
    One of the policeman approach to you and tells you:""")
    
    # Board Game
    print(f"""
    Policeman: {user_name}, this is you situation, think well before answering.
    \n
    {user_name} \ Computer\t\t Stay Silent \t\t\t  Snitch
    --------------\t --------------------------\t ---------------------------
    Stay Silent\t\t Both gets 3 years\t\t {user_name} gets 10 year, Python 1
    Snitch \t\t {user_name} gets 1 years, Python 10\t 25 years both
    """)
    
    # Create a board game with the pay-off matrix.
    board_game = [[3, 1], [5, 0]]
    
    # Save board points and results in a dictionary. R: remain silent, S: betray
    dictionary_results = {"(R, R)": (board_game[0][0], board_game[0][0]),
                          "(R, S)": (board_game[0][1], board_game[1][0]),
                          "(S, R)": (board_game[1][0], board_game[0][1]),
                          "(S, S)": (board_game[1][1], board_game[1][1])}
    
    # prompt the user to select what to do
    user_choose = input("""
    What would you do? Would you remain silent or cooperate with the police?
    Select:
    - 'R' to remain silent
    - 'S' to Snitch
    """).upper()
    
    # Computer (Python) chosee randomly either to remaint silent or betray
    computer_choose = random.randint(0,1) # 0: remain silent; 1: Betray
    
    # conditional structure with the outputs of each scenario
    if user_choose == "R" and computer_choose == 0:
        user_result = dictionary_results["(R, R)"][0]
        computer_result = dictionary_results["(R, R)"][1]
        print(f"""
    You've decided to remain silent, as well as me!
    You earned {user_result} points.
    I earned {computer_result}""")
    
    elif user_choose == "S" and computer_choose == 0:
        user_result = dictionary_results["(S, R)"][0]
        computer_result = dictionary_results["(S, R)"][1]
        print(f"""
    You've decided to snitch, I saved you!
    You earned {user_result} points.
    I earned {computer_result}""")

    elif user_choose == "R" and computer_choose == 1:
        user_result = dictionary_results["(R, S)"][0]
        computer_result = dictionary_results["(R, S)"][1]
        print(f"""
    I snitch you, you saved me!
    You earned {user_result} points.
    I earned {computer_result}""")

    elif user_choose == "S" and computer_choose == 1:
        user_result = dictionary_results["(S, S)"][0]
        computer_result = dictionary_results["(S, S)"][1]
        print(f"""
    We both snitched, we both go to jail
    You earned {user_result} points.
    I earned {computer_result}""")

# Declaring a winner.

    if user_result > computer_result:
        print(f"""Congrats, you won!""")

    elif user_result == computer_result:
        print(f"""Well we tie!""")

    elif user_result <= computer_result:
        print(f"""Congrats to me, I beat you!""")
    
if "no" in user_play:
    
    print(f"\n{user_name} We were caught! Do not confess! We didn't do anything!")
    
    print("""
    You are in small room at the police station. You can't communicate with anyone.
    One of the policeman approach to you and tells you:""")
    
    # Board Game
    print(f"""
    Policeman: {user_name}, this is you situation, think well before answering.
    \n
    {user_name} \ Computer\t\t Stay Silent \t\t\t  Snitch
    --------------\t --------------------------\t ---------------------------
    Stay Silent\t\t Both gets 3 years\t\t {user_name} gets 10 year, Python 1
    Snitch \t\t {user_name} gets 1 years, Python 10\t 25 years both
    """)
       
    def play_round(p1_strategy, p2_strategy):
    # player 1 chooses to remain silent or snitch
        p1_choice = p1_strategy()
    # player 2 chooses to remain silent or snitch
        p2_choice = p2_strategy()
    
    # set the payoffs for each outcome
        payoffs = {("R", "R"): (3, 3), ("R", "S"): (1, 5), ("S", "R"): (5, 1), ("S", "S"): (1, 1)}
        return payoffs[p1_choice, p2_choice]

    def random_strategy():
        return random.choice(["R", "S"])

    def tit_for_tat_strategy(last_choice=None):
        return random.choice(["R", "S"])
        if last_choice is None:
            return "R" 
            #return "R"
        return last_choice

    p1_score = 0
    p2_score = 0
    p1_strategy = tit_for_tat_strategy
    p2_strategy = random_strategy

    # play the game for 5 rounds
    for i in range(5):
        p1_result, p2_result = play_round(p1_strategy, p2_strategy)
        p1_score += p1_result
        p2_score += p2_result
        p1_strategy = tit_for_tat_strategy
        p2_strategy = random_strategy

    print(f"{user_name} score: {p1_score}")
    print(f"Python score: {p2_score}")

    # Declaring a winner.

    if p1_score > p2_score:
        print(f"""\nCongrats, {user_name} won!""")

    elif p1_score == p2_score:
        print(f"""\nWell we tie!""")

    elif p1_score <= p2_score:
        print(f"""\nPython won""")
```
